```python
import sys
def digiter(baseNum, counter):
    counter = counter + 1
    strNum = str(baseNum)
    arrNum = []
    for digit in strNum:
        arrNum.append(int(digit))

    increaseDig = sorted(arrNum, key=int)
    decreaseDig = sorted(arrNum, key=int, reverse = true)

#     print("Decreasing", decreaseDig)
#     print("increasing: ", increaseDig)
    numLarge = int(''.join(map(str,decreaseDig)))
    numSmall = int(''.join(map(str,increaseDig)))

    diff = numLarge-numSmall
    print("Round ", counter, ": ", numLarge, " - ", numSmall, " = ", diff)
    if(diff == 6174):
        return 0
    digiter(diff, counter)
digiter(2349, 0)
```
