```python
def decreasing_digits(n):
    a = n.digits()
    a.sort()
    while len(a) < 4:
        a.insert(0,0)
    return sum(i * (10^e) for e, i in enumerate(a))
def increasing_digits(n):
    a = n.digits()
    a.sort()
    a.reverse()
    return sum(i * (10^e) for e, i in enumerate(a))


for i in [1..10000]:#Running the Process 10 times. You should see a way to to write this as a while loop
    while(i != 0 and i != 1111 and i != 2222 and i != 3333 and i != 4444 and i != 5555 and i != 6666 and i != 7777 and i != 8888 and i != 9999):
        print decreasing_digits(i), '-' , increasing_digits(i), '=',decreasing_digits(i) - increasing_digits(i)
        i = decreasing_digits(i) - increasing_digits(i)
```
