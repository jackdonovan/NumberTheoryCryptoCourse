```python
'Test Entries in the Kernel of M'
a_prod = 1
c_prod = 1
for i in range(len(list_of_ws[0])):#Testing the First Entry!
    if list_of_ws[0][i] == 1:
        a_prod = a_prod*list_of_as[i]
        c_prod = c_prod*list_of_cs[i]
b_prod = sqrt(c_prod)
print a_prod
print b_prod
gcd(a_prod - b_prod,N)
```
